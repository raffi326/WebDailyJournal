**LATIHANTAHUN KABISAT**

```python
#untuk menghitung tahun kabisat
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)


test_years = [1800, 1900, 2000, 2020, 2024, 2100]
print("Tahun | Kabisat")
print("----------------")
for year in test_years:
    print(f"{year}  | {'Ya' if is_leap_year(year) else 'Tidak'}")
```

**LATIHAN BMI**

```python
# Praktikum: Menghitung BMI dan Kategori Berat Badan
def calculate_bmi(weight, height):
    if not (1.0 <= height <= 2.5 and 20 <= weight <= 200):
        return "Data tidak valid"

    bmi = weight / (height ** 2)

    if bmi < 18.5:
        category = "Kurus"
    elif bmi < 24.9:
        category = "Normal"
    elif bmi < 29.9:
        category = "Overweight"
    else:
        category = "Obesitas"

    return f"BMI: {bmi:.2f}, Kategori: {category}"

test_cases = [(50, 1.6), (70, 1.75), (90, 1.8), (110, 1.7)]

for weight, height in test_cases:
    print(f"Berat: {weight} kg, Tinggi: {height} m -> {calculate_bmi(weight, height)}")
```
